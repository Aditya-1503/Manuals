```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'salary': [1.7, 2.4, 2.3, 3.1, 3.7, 4.2, 4.4, 6.1, 5.4, 5.7, 6.4, 6.2],
        'experience': [1.2, 1.5, 1.9, 2.2, 2.4, 2.5, 2.8, 3.1, 3.3, 3.7, 4.2, 4.4]}

df = pd.DataFrame(data)
```

```python
df
```

```python
def calc_mse(b1, b0, X, Y):
    y_pred = b0 + b1*X
    mse = np.mean((y_pred-Y)**2)
    return mse
```

```python
b = 1.1
beta = [0.1, 0.8, 1.5]

Y = df['salary']
X = df['experience']
MSEs = []
for b1 in beta:
    MSEs.append(calc_mse(b1, b, X, Y))
```

```python
MSEs
```

```python
plt.plot(beta, MSEs)
plt.xlabel('slope')
plt.ylabel('MSE')
plt.show()
```

```python
mse = []
beta = []
b1 = 0
while b1 <= 1.5:
    mse.append(calc_mse(b1, b, X, Y))
    beta.append(b1)
    b1 += 0.01
```

```python
plt.plot(beta, mse)
plt.xlabel('slope')
plt.ylabel('MSE')
plt.show()
```
