```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mh = np.array([58, 62, 60, 64, 67, 70])
dh = np.array([60, 60, 58, 60, 70, 72])
df = pd.DataFrame({'Mother':mh, 'Daughter': dh})
```

```python
df
```

```python
class LinearRegression:
    def __init__(self,alp):
        self.b0 = 0
        self.b1 = 0
        self.alp = alp
    
    def predict(self, x):
        return self.b0 + self.b1*x
    
    def error(self, x, y):
        return np.mean((y - self.predict(x))**2)

    def iterate(self, x, y):
        self.b0 = self.b0 - self.alp*(-2*x*(y-self.predict(x)))
        self.b1 = self.b1 - self.alp*(-2*(y-self.predict(x)))
        
    
```

```python
mod = LinearRegression(0.0001)
```

```python
X, Y = df['Mother'], df['Daughter']
```

```python
loss = [[ _ for _ in range(500)], []]
for i in range(500):
    for x,y in zip(X,Y):
        mod.iterate(x, y)
    loss[1].append(mod.error(X,Y))
#         print(f' b0: {mod.b0}, b1 : {mod.b1}')
    print(mod.error(X,Y))
```

```python
# plt.scatter(X,Y, color='red')
plt.plot(loss[0], loss[1])
```
